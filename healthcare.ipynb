```python
import pandas as pd
import numpy as np
```

```python
df_csv = pd.read_csv("messy_data.csv")
```

```python
df_csv
```

```python
df_csv.info()
df_csv.isnull().sum()
df_csv.duplicated().sum()
```

```python
df_csv['Patient Name'] = df_csv['Patient Name'].str.title()
```

```python
df_csv
```

```python
pip install word2number
```

```python
from word2number import w2n
def convert_age(x):
    try:
        return int(x) 
    except:
        try:
            return w2n.word_to_num(x)  
        except:
            return None  

df_csv['Age'] = df_csv['Age'].apply(convert_age)
```

```python
df_csv
```

```python
df_csv['Age'] = df_csv['Age'].fillna(df_csv['Age'].median())
```

```python
df_csv
```

```python
df_csv.replace(["NaN", "none", "NONE", ""], np.nan, inplace=True)
```

```python
df_csv
```

```python
df_csv['Visit Date'] = df_csv['Visit Date'].astype(str).str.strip()
```

```python
df_csv
```

```python
df_csv['Visit Date'] = df_csv['Visit Date'].str.replace('.', '-', regex=False)
```

```python
df_csv
```

```python
df_csv['Visit Date'] = pd.to_datetime(
    df_csv['Visit Date'],
    errors='coerce',
    format='mixed'
)
```

```python
df_csv
```

```python
df_csv['Medication'] = df_csv['Medication'].replace(['NONE'], np.nan)
df_csv['Medication'] = df_csv['Medication'].fillna('No_Medication')
```

```python
df_csv
```

```python
df_csv[['systolic', 'diastolic']] = df_csv['Blood Pressure'].str.split('/', expand=True)

df_csv['systolic'] = pd.to_numeric(df_csv['systolic'], errors='coerce')
df_csv['diastolic'] = pd.to_numeric(df_csv['diastolic'], errors='coerce')
```

```python
df_csv
```

```python
df_csv.fillna({
    'systolic': df_csv['systolic'].median(),
    'diastolic': df_csv['diastolic'].median()
}, inplace=True)
```

```python
df_csv
```

```python
df_csv['Blood Pressure'] = df_csv['systolic'].astype(int).astype(str) + '/' + df_csv['diastolic'].astype(int).astype(str)
```

```python
df_csv
```

```python
df_csv['Email'] = df_csv['Email'].fillna('Unknown')
df_csv['Phone Number'] = df_csv['Phone Number'].fillna('Unknown')
```

```python
df_csv
```

```python
df_csv['Cholesterol'] = pd.to_numeric(df_csv['Cholesterol'], errors='coerce')
```

```python
print(df_csv['Cholesterol'].isna().sum())
print(df_csv['Cholesterol'].notna().sum())
```

```python
print(df_csv['Cholesterol'].dtype)
```

```python
df_csv['Cholesterol'].describe()
```

```python
df_csv['Cholesterol'].isna().mean() * 100
```

```python
df_csv['Cholesterol'] = df_csv['Cholesterol'].fillna(df_csv['Cholesterol'].median())
```

```python
df_csv.columns = df_csv.columns.str.capitalize()
```

```python
df_csv
```

```python
df_csv.to_csv("cleaned_data.csv", index=False)
```
