```python
import pandas as pd
import numpy as np

# Load dataset
df_csv = pd.read_csv("messy_data.csv")

# -------------------------------
# 1. Standardize column names
# -------------------------------
df_csv.columns = df_csv.columns.str.strip().str.lower().str.replace(" ", "_")

# -------------------------------
# 2. Handle missing values
# -------------------------------
# Replace 'Unknown' and empty strings with NaN
df_csv.replace(["Unknown", "", " "], np.nan, inplace=True)

# -------------------------------
# 3. Drop redundant column
# -------------------------------

# -------------------------------
# 4. Fix data types
# -------------------------------
# Convert age to numeric safely
df_csv["age"] = pd.to_numeric(df_csv["age"], errors="coerce")
df_csv["age"] = df_csv["age"].fillna(df_csv["age"].median()) # nullable integer
df_csv["age"] = df_csv["age"].astype("Int64")  # nullable integer
df_csv["visit_date"] = pd.to_datetime(df_csv["visit_date"], errors="coerce")

# -------------------------------
# 5. Clean categorical columns
# -------------------------------
# Standardize gender
df_csv["gender"] = df_csv["gender"].str.strip().str.lower()

# Standardize medication
df_csv["medication"] = df_csv["medication"].str.strip().str.upper()

# Replace NO_MEDICATION with NaN (optional)
df_csv["medication"] = df_csv["medication"].replace("NO_MEDICATION", np.nan)

# -------------------------------
# 6. Handle missing critical values
# -------------------------------
# Drop rows where condition is missing (important for ML)
df_csv = df_csv.dropna(subset=["condition"])

# -------------------------------
# 7. Handle duplicates
# -------------------------------
# Since no unique ID, remove exact duplicate rows
df_csv = df_csv.drop_duplicates()

# Optional: remove duplicates based on key fields
df_csv = df_csv.drop_duplicates(subset=["patient_name", "visit_date"], keep="first")

# -------------------------------
# 8. Validate numeric columns
# -------------------------------

df_csv["cholesterol"] = pd.to_numeric(df_csv["cholesterol"], errors="coerce")

# -------------------------------
# 9. Handle missing contact info (optional)
# -------------------------------
# Drop columns if not needed
df_csv.drop(columns=["email", "phone_number"], inplace=True, errors="ignore")

# -------------------------------
# 10. Reset index
# -------------------------------
df_csv.reset_index(drop=True, inplace=True)

# -------------------------------
# Final cleaned dataset
# -------------------------------
print(df_csv.info())
print(df_csv.head())

```
