# Patient Clustering — K-Means (K=4)
Features: `age`, `cholesterol`, `systolic`, `diastolic`

```python
# ─────────────────────────────────────────────
# IMPORTS (consolidated — no duplicates)
# ─────────────────────────────────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

# Load cleaned data

```python
# ─────────────────────────────────────────────
# LOAD DATA
# ─────────────────────────────────────────────
df = pd.read_csv('cleaned_data.csv')
df
```

```python
# ─────────────────────────────────────────────
# 1. FEATURES
# ─────────────────────────────────────────────
features = ['age', 'cholesterol', 'systolic', 'diastolic']
# Work on a copy to avoid mutating the original DataFrame on re-runs
df_out = df.copy()
X = df_out[features].values

# ─────────────────────────────────────────────
# 2. SCALING
# ─────────────────────────────────────────────
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ─────────────────────────────────────────────
# 3. K-MEANS  (K=4, justified below with silhouette)
# ─────────────────────────────────────────────
kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=50
)

df_out['cluster'] = kmeans.fit_predict(X_scaled)
centers = kmeans.cluster_centers_

# ─────────────────────────────────────────────
# 4. SILHOUETTE SCORE  (validates K=4)
# ─────────────────────────────────────────────
sil = silhouette_score(X_scaled, df_out['cluster'])
print(f"Silhouette score (K=4): {sil:.4f}")
print("  (Higher is better; range −1 to +1)")

# ─────────────────────────────────────────────
# 5. INVERSE TRANSFORM CENTROIDS  (original scale)
# ─────────────────────────────────────────────
centroids = scaler.inverse_transform(centers)
centroid_df = pd.DataFrame(centroids, columns=features)

print("\n📊 Cluster Centroids (Original Scale):")
print(centroid_df)

# ─────────────────────────────────────────────
# 6. CLUSTER SUMMARY
# ─────────────────────────────────────────────
print("\n📊 Cluster Mean Summary:")
cluster_mean = df_out.groupby('cluster')[features].mean()
print(cluster_mean)

global_mean = df_out[features].mean()
global_std  = df_out[features].std()

# ─────────────────────────────────────────────
# 7. PCA VISUALIZATION
# ─────────────────────────────────────────────
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
centers_pca = pca.transform(centers)

colors = cm.tab10(range(4))

plt.figure(figsize=(8, 6))

for i in range(4):
    mask = df_out['cluster'] == i
    plt.scatter(
        X_pca[mask, 0],
        X_pca[mask, 1],
        c=[colors[i]],
        label=f"Cluster {i}",
        s=40,
        alpha=0.7
    )

plt.scatter(
    centers_pca[:, 0],
    centers_pca[:, 1],
    c='black',
    marker='X',
    s=200,
    label='Centroids'
)

plt.title("K-Means Clustering (K=4)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend()
plt.tight_layout()
plt.show()

# ─────────────────────────────────────────────
# 8. CLUSTER DISTRIBUTION
# ─────────────────────────────────────────────
print("\n📊 Cluster Distribution:")
print(df_out['cluster'].value_counts().sort_index())
```

```python
# ─────────────────────────────────────────────
# 9. CLUSTER INTERPRETATION
#    Risk score is z-score based → thresholds are
#    statistically meaningful (SDs above global mean)
# ─────────────────────────────────────────────
print("\n🏥 FINAL CLUSTER INTERPRETATION\n")

for i in range(4):
    c = cluster_mean.loc[i]

    # Z-score based risk: how many SDs above global mean
    risk_score = (
        (c['systolic']    - global_mean['systolic'])    / global_std['systolic'] +
        (c['diastolic']   - global_mean['diastolic'])   / global_std['diastolic'] +
        (c['cholesterol'] - global_mean['cholesterol']) / global_std['cholesterol'] * 0.5
    )

    if risk_score > 0.5:
        label = "High-risk group"
    elif risk_score > 0:
        label = "Moderate-risk group"
    else:
        label = "Low-risk group"

    print(f"Cluster {i}  (risk_score={risk_score:.3f}):")
    print(f" → {label}\n")
```

# Save cluster results

```python
df_out.to_csv('Datamining/cluster_results.csv', index=False)
print(" cluster_results.csv saved")
```
